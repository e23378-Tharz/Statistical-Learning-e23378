# Q2) Bertrand's Paradox

# A chord of a circle is chosen "at random." What is the probability that its length exceeds the side length of the inscribed equilateral triangle.

# Show that different mathematically natural models of "random chord" lead to different answers.


1.The "Random Endpoints" Model
method:Pick two points at random on the circumference of the circle and draw a line between them.fix one point as a vertex of our inscribed triangle. To make the chord longer than the side of the triangle, the second point must land on the arc between the other two vertices of the triangle.      

Probability: $1/3 \approx 0.33$



2.The "Random Radial Position" Model
method:Pick a random radius of the circle, then pick a random point on that radius. Draw a chord perpendicular to the radius at that point.choose distance d from the center(from $0$ to $R$).
For a chord to be longer than the side of the inscribed triangle, it must be closer to the center than the midpoint of the radius(<=R/2)



Probability: $\frac{R/2}{R} = 1/2 = 0.50$



3.The "Random Midpoint" Model
Method:Pick any point at random anywhere inside the disk and treat it as the midpoint of a chord.Every point inside the circle (except the center) uniquely defines one chord.For the chord to be longer than the triangle side, its midpoint must fall within a smaller, inner circle that has a radius of $R/2$

Probability: $\frac{\pi(R/2)^2}{\pi R^2} = \frac{1/4 \pi R^2}{\pi R^2} = 1/4 = 0.25$



Conclusion: Bertrand's Paradox proves that "Uniform Distribution" is not a property of a set, but a property of a measure. Without specifying how you are choosing the chord, the question "What is the probability?" has no single correct answer.

# **Q10) Information Available in Continuous Random Varaibles**


**1.The Probability Space $(\Omega, \mathcal{F}, P)$**

The probability space is defined as:Sample Space: $\Omega = \mathbb{R}^2$.$\sigma$-algebra: $\mathcal{F} = \mathcal{B}(\mathbb{R}^2)$, which is the Borel $\sigma$-algebra on the plane.Probability Measure: For any $A \in \mathcal{F}$:$$P(A) = \iint_A f_{T,P}(t,p) \, dt \, dp$$where $f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$


**2 & 3. $\sigma$-algebras Generated by $X$ and $Y$**

$\sigma(X)$: This is the collection of events determined solely by temperature.$$\sigma(X) = \{ B \times \mathbb{R} : B \in \mathcal{B}(\mathbb{R}) \}$$$\sigma(Y)$: Since $Y$ is the identity map $Y(\omega) = \omega$, it generates the entire Borel $\sigma$-algebra.$$\sigma(Y) = \mathcal{B}(\mathbb{R}^2)$$

**4.Information Inclusion**

Because every set of the form $B \times \mathbb{R}$ is a Borel set in $\mathbb{R}^2$, we have:$$\sigma(X) \subset \sigma(Y)$$Explanation: $Y$ provides the coordinates for both temperature and pressure. $X$ only provides the first coordinate. Knowing $Y$ allows you to reconstruct $X$ perfectly, but knowing $X$ leaves the pressure $P$ completely unknown.

**5 & 6. Marginal Densities**

Marginal density of $X$: Since $X = T \sim N(0,1)$:$$f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$$Marginal density of $Y$: Since $Y = (T, P)$, its density is the joint density:$$f_Y(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$$

**7.Probability Computations**

$P(X \le 0) = \int_{-\infty}^0 f_X(x) \, dx = \mathbf{0.5}$

$P(X > 1) = 1 - \Phi(1) \approx 1 - 0.8413 = \mathbf{0.1587}$$P(Y \in (-\infty, 0] \times \mathbb{R}) = P(T \le 0) = \mathbf{0.5}$$P(Y \in [-1,1] \times [-2,2]) = P(|T| \le 1) \cdot P(|P| \le 2)$. Since $P \sim N(0, 4)$, $P/2 \sim N(0,1)$:$$P(|T| \le 1) \cdot P(|P/2| \le 1) = (0.6827) \cdot (0.6827) \approx \mathbf{0.466}$$

**8 & 9. Conditional Distributions**

Given $Y=(t,p)$: The temperature is fixed at $t$. The distribution is the Dirac delta measure:$$X | Y=(t,p) \sim \delta_t$$Given $X=t$: Because $T$ and $P$ are independent, knowing $T=t$ tells us nothing about $P$:$$Y | X=t \sim \{t\} \times N(0, 4)$$


**11.Measurability**

Define the projection $\pi_1(t,p) = t$. Then:$$X = \pi_1 \circ Y$$Since $\pi_1$ is a continuous mapping, it is Borel-measurable. In measure theory, if $X = g(Y)$ for a measurable $g$, then $X$ is $\sigma(Y)$-measurable, meaning all information in $X$ is contained within $Y$.

**13, 14, 15. Differential Entropy**

Using the formula for Gaussian entropy $h = \frac{1}{2} \ln(2\pi e \sigma^2)$:$h(X)$: $\frac{1}{2} \ln(2\pi e) \approx \mathbf{1.4189 \text{ nats}}$$h(Y)$: Due to independence, $h(Y) = h(T) + h(P)$:$$h(Y) = \frac{1}{2} \ln(2\pi e) + \frac{1}{2} \ln(2\pi e \cdot 4) = \ln(2\pi e) + \ln(2) \approx \mathbf{3.531 \text{ nats}}$$$h(Y|X)$: $h(Y,X) - h(X) = h(Y) - h(X) = h(P)$.$$h(Y|X) = \frac{1}{2} \ln(8\pi e) \approx \mathbf{2.112 \text{ nats}}$$Physical Interpretation: This represents the average uncertainty remaining about the reactor's state (the pressure fluctuations) once the temperature is known.


**16.Discussion of $h(X|Y)$**

For continuous variables, if $X$ is a deterministic function of $Y$:$$h(X|Y) \to -\infty$$This signifies that there is zero volume of uncertainty; the conditional distribution is a singular measure (a point), and its "spread" is zero.
